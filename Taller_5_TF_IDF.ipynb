```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances
```

```python
import nltk
nltk.download('all')
```

```python
stopwords_sp = stopwords.words('spanish')
```

# Punto 1: Pre-Procesamiento

- `[9 pts]` Leer el archivo `bob_esponja.csv` usando `pandas`
- `[9 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números, sin palabras vacias y hacer stemming de las palabras

```python
#Leer datos
datos = pd.read_csv('bob_esponja.csv', sep = ",")
datos
```

```python
# Nueva columna
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords_sp]
    texto = " ".join(texto)
    return texto

datos['pre-procesado'] = datos['Descripción'].apply(lambda texto: pre_procesado(texto))

datos
```

# Punto 2: TF-IDF

- `[16 pts]` Crear la matriz TF-IDF

```python
datos['pre-procesado'].values
```

```python
tfidf_vector = TfidfVectorizer()
tfidf = tfidf_vector.fit_transform(datos['pre-procesado'].values)

tfidf.toarray()


```

```python
tfidf_matrix = pd.DataFrame(tfidf.toarray())
tfidf_matrix.columns = tfidf_vector.get_feature_names()
tfidf_matrix.index = datos.index

tfidf_matrix.T
```

# Punto 3: Distancia del coseno
- `[12 pts]` Calcular la distancia del coseno entre cada una de los personajes
- `[2 pts]` ¿Cuáles son los personajes más parecidos?
- `[2 pts]` ¿Cuáles son los personajes más diferentes?

```python
dist_cos = cosine_distances(tfidf_matrix.values)
dist_cos = pd.DataFrame(dist_cos, columns = tfidf_matrix.index, index = tfidf_matrix.index)
dist_cos
```

# ¿Cuáles son los personajes más parecidos?

```python
Menos_parecidos = dist_cos[dist_cos == np.amax(dist_cos.values)] # Obtener el mayor valor en la matriz
Menos_parecidos
```

Los personajes menos parecidos son la Sra. Puff y Sheldon J. Plankton con un valor de 0.991454

# ¿Cuáles son los personajes más diferentes?

```python
dist_cos.values[dist_cos.values==0]=2
Mas_parecidos = dist_cos[dist_cos == np.amin(dist_cos.values)]# Buscar el valor menor en la matriz
Mas_parecidos
```

Los personajes más parecidos son Eugenio H. Cangrejo y Sheldon J. Plankton con un valor de 0.704685

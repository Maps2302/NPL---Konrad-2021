```python
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")
chachedWords = stopwords.words('english')
```

```python
stop = set(stopwords.words('english'))
```

```python
#Cargar los datos
data = pd.read_csv('/content/Base.csv', sep = ",")
data
```

```python
palabras = data.No_Stop1.str.cat(sep=" ")
wordcloud = WordCloud(width=800, height=400,
                      max_font_size=150, max_words=100,
                      background_color="white", colormap="seismic",
                      stopwords=stop, collocations=True).generate(palabras) 
                      
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

# Wordcloud con el Símbolo Republicano 

Cuenta la historia que el símbolo republicano del "elefante" nace como una crítica del caricaturista Thomas Nast, quien para el año 1874 publicó su obra titulada "Third Term Panic". En ella ilustra al como un enorme elefante, llamado «El voto republicano», que parece estar a punto de caer por un precipicio.

```python
img = PIL.Image.open("/content/Simbolo republicano.jpg")
display(img)
```

```python
mask = np.array(img)
mask
```

```python
palabras = data.No_Stop1.str.cat(sep=" ")
wordcloud = WordCloud(width=800, height=400,
                      max_font_size=150, max_words=100,
                      background_color="white", colormap="seismic",
                      stopwords=stop, collocations=True,
                      mask=mask, contour_width=0.5, contour_color='red').generate(palabras)


plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
              
```

# WordCloud con el Símbolo Demócrata

Siguiendo la historia de la caricatura titulada "Third Term Panic", Thomas Nast ilustró de igual manera al Partido Demócrata como un burro envuelto en la piel de un león, asustando a los otros animales con historias salvajes de una dictadura de Grant, mostrándolo como una criatura débil y asustada que constantemente avanzaba pesadamente en la dirección equivocada, su tamaño era más una responsabilidad que un activo.

```python
imagen = PIL.Image.open("/content/Simbolo democrata.jpg")
display(imagen)
```

```python
mascara = np.array(imagen)
mascara
```

```python
palabras = data.No_Stop1.str.cat(sep=" ")
wordcloud = WordCloud(width=800, height=400,
                      max_font_size=150, max_words=100,
                      background_color="white", colormap="seismic",
                      stopwords=stop, collocations=True,
                      mask=mascara, contour_width=1, contour_color='blue').generate(palabras)


plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
